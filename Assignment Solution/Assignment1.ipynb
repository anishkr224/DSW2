###### 1. Differentiate between the followings with proper examples
 
######  a. all() vs any()

all() function returns True if all elements in an iterable are true, otherwise it returns False.

Example:

```python
x = [True, True, False, True]
y = [1, 2, 3, 4]
print(all(x))  
print(all(y))
```

any() function returns True if any element in an iterable is true, otherwise it returns False.

Example:

```python
x = [False, False, True, False]
y = [0, '', False, None]
print(any(x))  
print(any(y)) 
```

###### b. dictionary vs default dictionary

dictionary is a built-in data type in Python used to store key-value pairs. Each key is unique and mapped to a value.

Example:

```python
d = {'a': 1, 'b': 2, 'c': 3}
print(d['a'])  
```

defaultdict is a subclass of the built-in dict class in Python that returns a default value if a key is not present.

Example:

```python
from collections import defaultdict
d = defaultdict(int)
print(d['a']) 
```

###### c. *args and **kwargs

*args is used to pass a variable number of non-keyworded arguments to a function. The arguments are passed as a tuple.

Example:

```python
def fun(*args):
    for arg in args:
        print(arg)
fun(1, 2, 3)
```

**kwargs is used to pass a variable number of keyworded arguments to a function. The arguments are passed as a dictionary.
 
Example:

```python
def fun(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
fun(a=1, b=2, c=3)
```

###### 2. Consider the data containing the salary and tenure of some employees. 

######  salaries_and_tenures=[(83000,8.7), (88000, 8.1), (48000, 0.7), (76000, 6), (69000, 6.5), (76000, 7.5), (89000, 8.7), (60000, 1.8), (83000, 3.5), (68000, 8.1), (48000, 0.7), (63000, 1.8), (25000, 3.5)]
 
###### a. Find the average salary of each tenure

```python
from collections import defaultdict

salaries_and_tenures = [(83000, 8.7), (88000, 8.1), (48000, 0.7), (76000, 6),
                        (69000, 6.5), (76000, 7.5), (89000, 8.7), (60000, 1.8),
                        (83000, 3.5), (68000, 8.1), (48000, 0.7), (63000, 1.8),
                        (25000, 3.5)]

# Create a defaultdict to group salaries by tenure
salaries_by_tenure = defaultdict(list)
for salary, tenure in salaries_and_tenures:
    salaries_by_tenure[tenure].append(salary)
    
print(salaries_by_tenure,'\n')

# Calculate average salary for each tenure
average_salaries_by_tenure = {}
for tenure, salaries in salaries_by_tenure.items():
    average_salaries_by_tenure[tenure] = sum(salaries) / len(salaries)
    
for tenure, average_salary in average_salaries_by_tenure.items():
    print('Tenure:', tenure, 'years, Average Salary: $',average_salary)
```

###### b. Group together the salaries corresponding to the following bucket of tenures, less than two, between two and five, more than 5
 


```python
buckets = {
    'Less than two': [],
    'Between two and five': [],
    'More than five': []
}

for salary, tenure in salaries_and_tenures:
    if tenure < 2:
        buckets['Less than two'].append(salary)
    elif 2 <= tenure <= 5:
        buckets['Between two and five'].append(salary)
    else:
        buckets['More than five'].append(salary)

for bucket, salaries in buckets.items():
    print('Salaries in',bucket,'tenure bucket:', salaries)
```

###### c. Compute the average salary for each group(bucket of tenures)

```python
average_salaries_by_bucket = {}
for bucket, salaries in buckets.items():
    average_salaries_by_bucket[bucket] = sum(salaries) / len(salaries)
    
for bucket, average_salary in average_salaries_by_bucket.items():
    print('Average Salary in', bucket, ': $',average_salary)
```

 ###### 4. Count frequencies of various alphabets (Convert upper case into lower case and input given by user), plot the results for this as a bar chart with x-axis being the letter and y-axis as the corresponding frequency.

```python
import matplotlib.pyplot as plt
from collections import Counter
#import pandas as pd

text = str(input('Enter String: '))
lowerCase = text.lower()

alphabet = []
for char in lowerCase:
    if char.isalpha():
        alphabet.append(char)

count = Counter(alphabet)
print(count)


letters = list(count.keys())
frequencies = list(count.values())

# print('Letter',letters)
# print('Frequency',frequencies)

plt.figure(figsize=(8, 6))
plt.bar(letters, frequencies)
plt.xlabel('Letter')
plt.ylabel('Frequency')
plt.title('Alphabet Frequency Bar Chart')
plt.show()
```

###### 7. Calculate total sale data for last year for each product and show it using a Pie chart Note: In Pie chart display Number of units sold per year for each product in percentage.

```python
import pandas as pd
import matplotlib.pyplot as plt

products = df.columns[1:7]  # Get all product columns except
plt.figure(figsize=(8, 8))

data=[]
for product in products:
    total_sales_last_year = df[product].sum()
    data.append(total_sales_last_year)
    
plt.pie(data, labels=products, autopct='%1.1f%%')
plt.title('Total Sales for Last Year by Product')
plt.show()
```
